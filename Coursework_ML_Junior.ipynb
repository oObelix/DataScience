# 2. Исследование зависимостей и формулирование гипотез

```python
# Создадим матрицу корреляций
corr_m = df_users.corr()
```

```python
# Отобразим матрицу корреляций
plt.figure(figsize=(24, 10))
sns.heatmap(corr_m, center=0, annot=True)
plt.show()
```

```python
# Выбреем для примера поля, которые наиболее коррелируют с 'Churn' по модулю.
# Для этого отфильтруем кореляции меньшие по модулю чем 0.17, что даст нам набор 
# из 8 фичей, который возьмём за максимум, за гранией которого мы будем иметь 
# однозначно переобученную модель

corr_min = 0.17
corr_m['Churn'][abs(corr_m['Churn']) >= corr_min].drop('Churn')
```

```python
# Для проверки сравним с тем, как автоматические страдства 
# sklearn.feature_selection выберет наиболее коррелирующие поля фичей.
# Видно, что набор почти тот же самый.
 
from sklearn.feature_selection import mutual_info_regression
 
X = df_users.drop(columns=['customerID', 'Churn'])
y = df_users['Churn']
 
mi_scores = mutual_info_regression(X, y, discrete_features=True)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)
 
mi_scores
```

# 3. Построение моделей для прогнозирования оттока

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import recall_score

y = df_users['Churn']
# X = df_users.drop(columns=['tenure', 'MonthlyCharges', 'TotalCharges', 'PaymentMethod', 'customerID', 'Churn'])
# X = df_users[features.drop('Churn')]
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Создадим функцию, которая будет принимать на вход минимальное значение 
# кореляции и обучать с соответствующим набором фичей сразу несколько моделей,
# с последующим выводом результата в виде таблицы метрик.
# В качестве моделей возьмём все, которые проходили в теоретической части.
# В качестве метрики качества обучения будем использовать ROC_AUC и Recall.
# Recall нужен для наиболее полного охвата оттока.
# Accuracy тут скорее формальна, так как набор данных не сбалансирован.

def classifiers(corr_min = 0.17, features=[], drop_columns=[]):
  if not features:
    features = corr_m['Churn'][abs(corr_m['Churn']) >= corr_min].index
    features = features.drop('Churn')
  
  X = df_users[features]

  if drop_columns:
    X = X.drop(drop_columns, axis=1)

  print('Features:', list(X))
  print('Total Features:', len(X.columns))

  X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

  dtc = DecisionTreeClassifier(random_state=0)
  dtc_model = dtc.fit(X_train, y_train)

  nb = GaussianNB()
  nb_model = nb.fit(X_train, y_train)

  knn = KNeighborsClassifier(n_neighbors=10)
  knn_model = knn.fit(X_train, y_train)

  lr = LogisticRegression(max_iter=2000, random_state=0)
  lr_model = lr.fit(X_train, y_train)

  accuracy = dtc_model.score(X_test, y_test)
  roc_auc = roc_auc_score(y_test, dtc_model.predict_proba(X_test)[:, 1])
  recall = recall_score(y_test, dtc_model.predict(X_test))
  print('DecisionTree \t Accuracy: %.4f,  ROC_AUC: %.4f,  Recall: %.4f' % (accuracy, roc_auc, recall))

  accuracy = nb_model.score(X_test, y_test)
  roc_auc = roc_auc_score(y_test, nb_model.predict_proba(X_test)[:, 1])
  recall = recall_score(y_test, nb_model.predict(X_test))
  print('GaussianNB \t Accuracy: %.4f,  ROC_AUC: %.4f,  Recall: %.4f' % (accuracy, roc_auc, recall))

  accuracy = knn_model.score(X_test, y_test)
  roc_auc = roc_auc_score(y_test, knn_model.predict_proba(X_test)[:, 1])
  recall = recall_score(y_test, knn_model.predict(X_test))
  print('KNeighbors \t Accuracy: %.4f,  ROC_AUC: %.4f,  Recall: %.4f' % (accuracy, roc_auc, recall))

  accuracy = lr_model.score(X_test, y_test)
  roc_auc = roc_auc_score(y_test, lr_model.predict_proba(X_test)[:, 1])
  recall = recall_score(y_test, lr_model.predict(X_test))
  print('LogisticRegr \t Accuracy: %.4f,  ROC_AUC: %.4f,  Recall: %.4f' % (accuracy, roc_auc, recall))
```

```python
classifiers(corr_min = 0.17)
```

```python
classifiers(corr_min = 0.18)
```
