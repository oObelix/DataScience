# pd.plotting.**scatter_matrix** на примере load_iris датасета.

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
import pandas as pd
```

```python
iris_dataframe = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
iris_dataframe
```

```python
iris_dataset.target
```

```python
scat_mtrx = pd.plotting.scatter_matrix(iris_dataframe, figsize=(16, 16), c=iris_dataset.target, marker='o', hist_kwds={'bins': 20}, s=40, alpha=.8)
```

```python
iris_dataframe_simple = pd.DataFrame(iris_dataset.data[:, 2:4], columns=iris_dataset.feature_names[2:4])
iris_dataframe_simple
```

```python
scat_mtrx = pd.plotting.scatter_matrix(iris_dataframe_simple, figsize=(10, 10), c=iris_dataset.target, marker='o')
```
