Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
# -- ВАШ КОД ТУТ --

import pickle
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
with open('sample_data/clustering.pkl', 'rb') as f:
  clustering = pickle.load(f)

X = np.array(clustering['X'])
# X
```

```python
db = DBSCAN(eps=0.5, min_samples=8).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_)
plt.show()
```

```python
db = DBSCAN(eps=0.6, min_samples=8).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_)
plt.show()
```

```python
db = DBSCAN(eps=0.7, min_samples=8).fit(X)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_)
plt.show()
```
