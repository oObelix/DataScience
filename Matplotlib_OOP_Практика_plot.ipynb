Используя объектно-ориентированный подход, постройте линейный график минимальной зарплаты в штате Аризона до 2000 года включительно. После этого отформатируйте график, сделав заголовок и подписи осей темно-синими.

```python
!ls sample_data
```

```python
df = pd.read_csv('sample_data/wage-data-coast.csv')
df
```

```python
data = df[(df['State'] == 'Arizona') & (df['Year'] <= 2000)]
data
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(data['Year'], data['Salary'])
ax.set_title('График минимальной зарплаты в штате Аризона до 2000г', pad=26, 
             size=20, c='Navy')
ax.set_ylabel('$/час', c='Navy')
x_range = list(range(data['Year'].min(), data['Year'].max() + 1, 2)) + [2000]
ax.set_xticks(x_range)
plt.show()
```
